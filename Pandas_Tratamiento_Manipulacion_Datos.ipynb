#**PROBLEMA DE NEGOCIO**


---




Se trata de un algoritmo capaz de analizar las características de una propiedad -como comodidades, tamaño, ocupación del espacio en un período determinado- y sugerir al anfitrión un precio a cobrar por tarifas diarias que garantice ganancias en momentos de alta demanda.

# #**1. Entendiendo el Problema**


---




```python
import pandas as pd
```

```python
datos = pd.read_json('/content/datos_hosting.json')
datos.head()
```

```python
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

#**2. Datos numericos**

```python
columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[3:])
datos
```

```python
datos.reset_index(inplace = True, drop =True)
datos.head()
```

```python
datos.info()
```

```python
import numpy as np
```

```python
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64) ##cambia el tipo de dato a a entero
```

```python
datos.info()
```

```python
col_numericas =['cantidad_baños', 	'cantidad_cuartos', 	'cantidad_camas']
```

```python
datos[col_numericas] = datos[col_numericas].astype(np.int64) ##cambia el tipo de dato a a entero
```

```python
datos.info()
```

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64) ##cambia el tipo de dato a decimal
```

```python
datos.info()
```

```python
datos['precio']= datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip()) #con este metodo replazara los signos encontrados por nada
```

```python
datos.info()
```
