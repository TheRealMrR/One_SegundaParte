# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape #ver cuantos registros tiene
```

```python
df.info()
```

```python
df.set_index('Pais',inplace = True) #Ordena la tabla segun el pais
df
```

```python
anos = list(map(str,range(1980,2014))) #lista con todos los anos que hay en la tabla
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

```python
col_dic = {'Ano' : colombia.index.tolist(),
           'Inmigrantes' : colombia.values.tolist()}
```

```python
datos_col = pd.DataFrame(col_dic)
datos_col.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Ano'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Ano'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigracion de Colombianos hacia Canada')
plt.xlabel('Ano')
plt.ylavel('Numero de Inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax.plot(datos_col['Ano'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigracion de colombianos hacia Canada \n1980 -2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize = (10,3))
axs[0].plot(datos_col['Ano'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigracion de colombianos hacia Canada \n1080 - 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()


axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplor de inmigracion de colombianos \nhacia Cnada 19080 -2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero Inmigrantes')
axs[1].grid()
plt.show()
```
